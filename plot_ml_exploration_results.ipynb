```python
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20,5))

x = [.1624, .0343, .1272, .1012, .1495]
y = [26, 4, 79, 6, 18]
labels = ['light gbm', 'ridge', 'ridge search 10', 'linear svm', 'xgb']


axs[0].scatter(x, y, alpha=.5)
axs[0].set_xlabel('r2')
axs[0].set_ylabel('Seconds')
axs[0].set_title('SST predicting SSRT')
axs[0].set_xlim(0,.2)


for i, txt in enumerate(labels):
    axs[0].annotate(txt, (x[i], y[i]))


x = [.16452, .1547, .1944, .1867, .1903, .1022]
y = [12, 2, 21, 70, 8, 25]
labels = ['light gbm', 'ridge', 'ridge search 10', 'rf', 'xgb', 'linear svm']

axs[1].scatter(x, y, alpha=.5)
axs[1].set_xlabel('r2')
axs[1].set_ylabel('Seconds')
axs[1].set_title('nBack predicting 0-back prime')
axs[1].set_xlim(.09,.23)

for i, txt in enumerate(labels):
    axs[1].annotate(txt, (x[i], y[i]))
    
x = [.0908, .10592,  .10222, .0982, .0967]
y = [10, 8, 73, 21, 13]
labels = ['light gbm', 'xgb',  'rf', 'ridge 10', 'ridge 5']

axs[2].scatter(x, y, alpha=.5)
axs[2].set_xlabel('r2')
axs[2].set_ylabel('Seconds')
axs[2].set_title('nBack predicting 2-back prime')

for i, txt in enumerate(labels):
    axs[2].annotate(txt, (x[i], y[i]))
    
plt.savefig('destr_quick.png')
```
